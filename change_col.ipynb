```python
import os
import mysql.connector
from mysql.connector import Error
from itertools import combinations
import time
import matplotlib.pyplot as plt
import numpy as np
from prettytable import PrettyTable
import statistics


def get_column_names(database, table_name):
    connection = mysql.connector.connect(user="root", password="yzh_20010315", database=database)
    cursor = connection.cursor()
    """Get all column names of a database table"""
    cursor.execute(f"DESCRIBE `{table_name}`")

    cols = [column[0] for column in cursor.fetchall()]

    if connection.is_connected():
        cursor.close()
        connection.close()

    return cols



def pmax_uniqueness_ratio_sql(cursor, table_name, columns):
    """Calculates the Uniqueness Ratio for a given column permutation"""
    if not columns:
        return 0, 0
    columns_list = ', '.join([f"`{col}`" for col in columns])
    query = f"""
        SELECT 
            ((null_rows + unique_rows) / total_rows) AS uniqueness_ratio
        FROM (
            SELECT
                (SELECT COUNT(*) FROM `{table_name}` WHERE {' OR '.join([f"`{col}` IS NULL" for col in columns])}) AS null_rows,
                (
                    SELECT COUNT(*)
                    FROM (
                        SELECT {columns_list}
                        FROM `{table_name}`
                        WHERE {' AND '.join([f"`{col}` IS NOT NULL" for col in columns])}
                        GROUP BY {columns_list}
                        HAVING COUNT(*) = 1
                    ) AS unique_data
                ) AS unique_rows,
                (SELECT COUNT(*) FROM `{table_name}`) AS total_rows
        ) AS results;
    """
    start_time = time.time()
    cursor.execute(query)
    result = cursor.fetchone()
    execution_time = (time.time() - start_time) * 1000 # Convert to milliseconds
    ratio = result[0] if result else 0

    return ratio, execution_time


# calculate the number of null values in a given column
def num_of_null(table, sub_column, database, row_count):
    connection = mysql.connector.connect(user='root', password='yzh_20010315', database=database)
    cursor = connection.cursor()
    if not sub_column:
        return 0, 0
    
    cursor.execute("DROP TABLE IF EXISTS TempTable;")
    cursor.execute(f"CREATE TABLE TempTable AS SELECT * FROM `{database}`.`{table}` LIMIT {row_count};")
    
    sub_column_list = ' OR '.join([f"`{col}` IS NULL" for col in sub_column])
    query = f"""
        SELECT COUNT(*) FROM TempTable WHERE {sub_column_list};
    """
    cursor.execute(query)
    result = cursor.fetchone()
    
    cursor.execute("DROP TABLE IF EXISTS TempTable;")

    if connection.is_connected():
        cursor.close()
        connection.close()

    return result


def calculation_sql(user, password, database, table_name, columns, row_count):
    try:
        connection = mysql.connector.connect(user=user, password=password, database=database)

        if connection.is_connected():
            cursor = connection.cursor()
            all_combinations = [combo for length in range(1, 11) for combo in
                                combinations(columns, length)]
            execution_times_by_length = {}
            detailed_output = []

            cursor.execute("DROP TABLE IF EXISTS TempTable;")
            cursor.execute(f"CREATE TABLE TempTable AS SELECT * FROM `{database}`.`{table_name}` LIMIT {row_count};")

            # Calculate the uniqueness ratio for each column combination
            for perm in all_combinations:
                ratio, exec_time = pmax_uniqueness_ratio_sql(cursor, "TempTable", perm)
                key = len(perm)

                if key in execution_times_by_length:
                    execution_times_by_length[key].append(exec_time)
                else:
                    execution_times_by_length[key] = [exec_time]
                detailed_output.append((perm, ratio, exec_time))

            cursor.execute("DROP TABLE IF EXISTS TempTable;")

            # Calculate average execution time by length of column combination
            avg_times = []
            lengths = []
            for length, times in sorted(execution_times_by_length.items()):
                avg_time = sum(times) / len(times)
                avg_times.append(avg_time)
                lengths.append(length)

    except Error as e:
        print("Error:", e)
    finally:
        if connection.is_connected():
            cursor.close()
            connection.close()

    return detailed_output, avg_times, lengths


# Fitting curve function
def fit_curve(lengths, times, degree=2):
    """Fit a polynomial curve of given degree to the lengths and times."""
    coefficients = np.polyfit(lengths, times, degree)
    polynomial = np.poly1d(coefficients)
    fitted_values = polynomial(lengths)
    SS_res = np.sum((times - fitted_values) ** 2)
    SS_tot = np.sum((times - np.mean(times)) ** 2)
    r_squared = 1 - (SS_res / SS_tot)
    
    return polynomial, coefficients, r_squared


def main_sql(num, table_name, columns, row_count, database):
    global lengths, avg_times
    averages = []
    total_time = []

    for i in range(1, num + 1):
        detail_out, avg_times, lengths = calculation_sql('root', 'yzh_20010315', database, table_name, columns, row_count)
        total_time.append(avg_times)
    num_columns = len(avg_times)

    for col_index in range(num_columns):
        column_values = [row[col_index] for row in total_time]

        # sorted column_values
        column_values.sort()

        # Calculate the average value after removing extreme values
        avg = sum(column_values) / len(column_values)
        averages.append(avg)

    # Fit the curve and calculate the coefficients
    polynomial, coefficients, r_square = fit_curve(lengths, averages, 1)
    formatted_coefficients = [f"{coef:.2f}" for coef in coefficients]
    print(f"Polynomial Coefficients: {formatted_coefficients}")
    print("Averages: ", averages)
    print(f"r_square: {r_square:.4f}")

    # draw graph
    fig, ax1 = plt.subplots(figsize=(14, 7))
    
    # Plot average execution time
    line1, = ax1.plot(lengths, averages, marker='o', linestyle='-', color='black', label='Average Execution Time')
    line2, = ax1.plot(lengths, polynomial(lengths), linestyle='--', color='b', label=f'Fitted Curve (degree={len(coefficients) - 1})')

    # Update font sizes
    ax1.set_xlabel('Number of Columns |X|', fontsize=24)
    ax1.set_ylabel('Time in Milliseconds (t)', color='black', fontsize=24)
    ax1.tick_params(axis='y', labelcolor='black', labelsize=24)
    ax1.tick_params(axis='x', labelsize=24)
    ax1.set_title(f'Runtime VS Columns \nTable Name: {table_name}', fontsize=24)
    for i in range(len(lengths)):
        ax1.text(lengths[i], averages[i], f'{averages[i]:.4f}', fontsize=24, ha='left', color='black')
    ax1.grid(True)

    # Adding the fitted equation to the plot
    equation_text = f"y = {' + '.join([f'{coef:.4f}x^{i}' for i, coef in enumerate(coefficients)])}"
    ax1.text(0.05, 0.95, equation_text, transform=ax1.transAxes, fontsize=20, color='blue', verticalalignment='top')

    fig.tight_layout()

    
    # Save the plot
    desktop_path = os.path.join(os.path.expanduser("~"), "Desktop/colRunTime")
    file_name = f"{table_name}Col.png"
    file_path = os.path.join(desktop_path, file_name)

    fig.tight_layout()
    plt.savefig(file_path, dpi=300)  # Save plot to desktop with high resolution
    plt.show()

    print(f"Plot saved successfully as {file_path}")


    return averages


    
l1 = {
    'Goalies': ['playerID', 'year', 'tmID', 'lgID', 'GP', 'Min', 'W', 'L', 'GA', 'stint'], 
    'Teams': ['year', 'tmID', 'name', 'lgID', 'franchID', 'playoff', 'confID', 'divID', 'GF', 'GA'],
    'TeamVsTeam': ['year', 'lgID', 'tmID', 'oppID', 'W', 'L', 'T', 'OTL'],
    'SeriesPost': ['year', 'tmIDWinner', 'tmIDLoser', 'round', 'series', 'lgIDWinner', 'lgIDLoser', 'GoalsWinner', 'GoalsLoser', 'note'], 
    'TeamsPost': ['year', 'tmID', 'lgID', 'W', 'G', 'PIM', 'GA', 'GF', 'L', 'T'],
    'Coaches': ['coachID', 'year', 'tmID', 'lgID', 'stint', 'g', 'w', 'l', 't', 'postg'],
    'Scoring': ['playerID', 'year', 'tmID', 'lgID', 'stint', 'pos', 'GP', 'G', 'A', 'postG'], 
    'TeamSplits': ['year', 'tmID', 'lgID', 'hW', 'hL', 'hOTL', 'rW', 'rL', 'rT', 'rOTL'], 
    'AwardsPlayers': ['playerID', 'award', 'year', 'lgID', 'note', 'pos'], 
    'Master': ['playerID', 'coachID', 'firstName', 'lastName', 'nameGiven', 'legendsID', 'ihdbID', 'hrefID', 'birthYear', 'birthCountry']
}
ten_average_time = []

for key, values in l1.items():
    print("--------------------------------------------------------------------------------------------------------------------")
    print(f"Table: {key}\nColumns: {values}")
    tt = main_sql(num=5, table_name=key, columns= values, row_count=7000, database='hockey')
    ten_average_time.append(tt)
    

print("ten_average_time: ", ten_average_time)
```
